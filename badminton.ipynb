```python
# Convert to string first, then clean
df['PRIZE MONEY'] = df['PRIZE MONEY'].astype(str).str.replace('$', '').str.replace(',', '')
df['PRIZE MONEY'] = pd.to_numeric(df['PRIZE MONEY'], errors='coerce')  # converts invalid to NaN
df['PRIZE MONEY'].fillna(0, inplace=True)  # fill NaN with 0
df['PRIZE MONEY'] = df['PRIZE MONEY'].astype(int)  # convert to int
```

```python
df.info()
```

```python
def calculate_win_rate(win_lose_str):
   if pd.isna(win_lose_str):
       return 0.0
   
   # Split by ' - ' to get wins and losses
   parts = str(win_lose_str).split(' - ')
   if len(parts) != 2:
       return 0.0
   
   try:
       wins = int(parts[0])
       losses = int(parts[1])
       total = wins + losses
       
       if total == 0:
           return 0.0
       
       return wins / total
   except:
       return 0.0
   
df['WIN_RATE'] = df['WIN - LOSE'].apply(calculate_win_rate)
df.drop('WIN - LOSE', axis=1, inplace=True)
```

```python
df.drop('CHANGE +/-', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Split the POINTS / TOURNAMENTS column and extract just points
def extract_points(points_tournaments_str):
   if pd.isna(points_tournaments_str):
       return 0
   
   # Split by ' / ' to get points and tournaments
   parts = str(points_tournaments_str).split(' / ')
   if len(parts) != 2:
       return 0
   
   try:
       points = int(parts[0].replace(',', ''))  # remove commas if any
       return points
   except:
       return 0

# Apply to create new POINTS column
df['POINTS'] = df['POINTS / TOURNAMENTS'].apply(extract_points)
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
mask = df['PRIZE MONEY'] > 0
df_clean = df[mask].copy()
# Remove extreme ranks (keep only top 300 players)
mask1 = (df_clean['RANK'] >= 1) & (df_clean['RANK'] <= 300)

# Remove outliers based on percentiles for other features
# Remove bottom/top 5% of WIN_RATE
win_rate_lower = np.percentile(df_clean['WIN_RATE'], 5)
win_rate_upper = np.percentile(df_clean['WIN_RATE'], 95)
mask2 = (df_clean['WIN_RATE'] >= win_rate_lower) & (df_clean['WIN_RATE'] <= win_rate_upper)

# Remove bottom/top 5% of POINTS
points_lower = np.percentile(df_clean['POINTS'], 5)
points_upper = np.percentile(df_clean['POINTS'], 95)
mask3 = (df_clean['POINTS'] >= points_lower) & (df_clean['POINTS'] <= points_upper)

# Combine all masks
df_ultra_clean = df_clean[mask1 & mask2 & mask3].copy()

print(f"Original size: {len(df_clean)}")
print(f"After removing outliers: {len(df_ultra_clean)}")
```

```python
df_ultra_clean.head(500)
```

```python
# Use all your cleaned features
X = df_ultra_clean[['PRIZE MONEY', 'WIN_RATE', 'POINTS']].values
y = df_ultra_clean['RANK'].values

# Scale everything to similar ranges (0-1)
X[:, 0] = X[:, 0] / np.max(X[:, 0])  # prize money to 0-1
X[:, 1] = X[:, 1]                    # win rate already 0-1
X[:, 2] = X[:, 2] / np.max(X[:, 2])  # points to 0-1

# Add bias column
X = np.column_stack([np.ones(len(X)), X])  # now 4 columns

# Reset weights for 4 features (bias + 3 features)
np.random.seed(42)
weights = np.random.normal(0, 0.01, 4)

print(f"New feature matrix shape: {X.shape}")
print(f"New weights shape: {weights.shape}")
```

```python
def predict(X, weights):
   return X @ weights
def cost_function(X, y, weights):
    predictions = predict(X, weights)
    errors = predictions - y
    mse = np.mean(errors ** 2)
    return mse
def compute_gradients(X, y, weights):
    m = len(y)  # number of samples
    predictions = predict(X, weights)
    errors = predictions - y
    gradients = (1/m) * X.T @ errors
    return gradients
def gradient_descent(X, y, weights, learning_rate, num_iterations):
    costs = []
    
    for i in range(num_iterations):
        # Calculate current cost
        cost = cost_function(X, y, weights)
        costs.append(cost)
        
        # Calculate gradients
        gradients = compute_gradients(X, y, weights)
        
        # Update weights
        weights = weights - learning_rate * gradients
        
        # Print progress every 100 iterations
        if i % 50 == 0:
            print(f"Iteration {i}, Cost: {cost:.2f}")
    
    return weights, costs

# Train the model
final_weights, cost_history = gradient_descent(X, y, weights, learning_rate=0.001, num_iterations=15000)

print(f"Final weights: {final_weights}")
print(f"Initial cost: {cost_history[0]:.2f}")
print(f"Final cost: {cost_history[-1]:.2f}")
```

```python
# Make predictions on your training data
predictions = predict(X, final_weights)
count = 227
avg_diff = 0

# Show some real examples
for i in range(count):
    
    actual_rank = y[i]
    pred_rank = predictions[i]
    avg_diff += actual_rank - pred_rank
    win_rate = df_ultra_clean.iloc[i]['WIN_RATE']
    points = df_ultra_clean.iloc[i]['POINTS']
    player_name = df_ultra_clean.iloc[i]['PLAYER']
    
    print(f"Player: {player_name}")
    print(f"Win Rate: {win_rate:.3f}, Points: {points}")
    print(f"Actual Rank: {actual_rank}, Predicted: {pred_rank:.1f}")
    print("-" * 40)
print("Average Difference")
print(avg_diff/count)
```
